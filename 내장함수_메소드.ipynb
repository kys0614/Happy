# 학생 점수를 입력받아 평균점수, 최고점수, 최저점수를 구하는
# 성적 프로그램을 구현하라!

# 점수를 입력하세요 더이상 입력할 점수가 없으면 음수를 아무거나 입력해라
# 점수입력>>>80
# 점수입력>>>50
# 점수입력>>>60
# 점수입력>>>30
# 점수입력>>>90
# 점수입력>>>-1

# 평균 = 62.0 최대 = 90 최소 = 30

```python
print("점수를 입력하세요. 더이상 입력할 점수가 없으면 음수를 아무거나 입력하세요")

exam = []
while True:
  score = int(input("점수 입력>>>"))

  if score < 0 :
    break
  else:
    exam.append(score)
avg = sum(exam)/len(exam)  #평균
#print(avg)
max_ = max(exam)  #최대값
min_ = min(exam)  #최소값

print("평균={} 최대={} 최소={}".format(avg,max_,min_))
```

# 리스트 메소드 : 리스트라는 객체를 지원해주는 메소드다.

# append : 리스트 오른쪽에 자동으로 추가되는 데이터를 만드는 메소드다.
# extend : 리스트에 다른 리스트를 확장해서 결합할 수 있는 메소드다.
# insert : 리스트에 위치값의 데이터를 추가할 수 있는 메소드다.
# clear : 리스트에 저장된 모든 요소를 삭제하는 메소드다.
# pop : 리스트의 마지막 요소를 반환할 때 사용하는 메소드다.
# remove : 리스트 해당요소를 지정('apple' : 값)을 해서 삭제하는 메소드다.

# 교집합 : & -> 새로운 집합끼리 묶어서 하나의 리스트를 만든다.
# 합집합 : | -> 모든 요소를 합친 집합을 생성하는 리스드를 만든다.
# 차집합 : - -> 다른 집합의 요소를 제거한 새로운 형태의 리스트를 만든다.

```python
s1 = {'apple','banana','cherry'}
s2 = {'apple','banana','orange'}
s1 & s2
s1 | s2
s2 - s1
```
